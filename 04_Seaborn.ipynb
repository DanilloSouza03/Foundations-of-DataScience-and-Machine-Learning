```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
%matplotlib inline
```

**DataFrame de Voos**

```python
df = pd.read_csv('dataframes/flights.csv', sep=',')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

**Inicializando com os plots**

```python
sns.displot(df['passengers'], bins=50)
```

```python
plt.figure(figsize=(12,8))
sns.barplot(x='year',y='passengers',data=df)
```

```python
sns.jointplot(x='year',y='passengers',data=df)
```

```python
sns.pairplot(df,hue='month')
```

```python
sns.heatmap(df.corr(), annot=True)
```
