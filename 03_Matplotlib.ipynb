```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
%matplotlib inline
```

**Primeiros plot**

```python
x = [5,10,15,20,25]
y = [15,25,45,55,60]

plt.plot(x,y)
plt.show()
```

```python
plt.bar(x,y)
plt.show()
```

```python
plt.barh(x,y)
plt.show()
```

```python
plt.scatter(x,y)
plt.show()
```

```python
x = [5,10,15,20,25]
y = [15,25,45,55,60]
fig, axs = plt.subplots(figsize=(10,4))
axs.set_xlabel('Eixo X')
axs.set_ylabel('Eixo Y')
axs.set_title('Título do Gráfico')
axs.plot(x,y, label='Dados 1')
axs.plot(y,x, label='Dados 2')
axs.legend()
```
